Regression:It is method for understanding relationship btw dependent and independent variables.

--it is used on continues dataset
--2 types of regression

1)Simple regression:If one feature is mapped for target then it is simple reg.

A)Linear simple regression:if this selected feature and target have linear relationship then it is called simple linear regression.

B)non-linear simple regression:if there is non-linearity btw features and target then it is simple non linear reg.

2)Multiple Regression:-If more than one feature is mapped for target then it is simple reg.


A)Linear multiple regression:if this selected feature and target have linear relationship then it is called multiple linear regression.

B)non-linear multiple regression:If more than one feature is mapped for target then it is multiple non-linear reg.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.array([1,2,3,4])
slope=0.5
intercept=4
y=slope*x+intercept
y
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
area=np.array([600,800,1000,1200,1500,2000])
price=np.array([4,6,9,12,16,21])
plt.scatter(area,price)
```

```python
ypred=area*slope+intercept
```

```python
area_m=np.mean(area)
price_m=np.mean(price)
slope=np.sum((area-area_m)*(price-price_m))/np.sum((area-area_m)**2)
intercept=price_m-(slope*area_m)
```
