We can fill missing values with 3 diff.values

skewness:

it is measure of asymmetry.It tells us how datapoits are distributed
type:-
1)+ve 2)-ve 3) zero
we reduce skewness by using 2 methods
1)longvalue 2)square root

```python
from scipy.stats import skew
```

```python
colname=X.select_dtypes(['int64','float64']).columns
colname
```

```python
#apply skew function on one column
skew(X["engine-size"])
```

```python
#apply loop for calculating skew for all col
for col in X[colname]:
  print(col)
  print(skew(X[col]))
  plt.figure()
  sns.distplot(X[col])
  plt.show()
```

Before removing skewness we need to check correlation of features with target
follow 2 rules
1)if the cloumn is skewd but showing good correlaton with target then dont remove skewness
2)if the col which has some negative values then dont skewness even through showing poor correlation with target

```python
#removing skewness
X['engine-size']=np.log(X['engine-size'])
skew(X['engine-size'])
```
